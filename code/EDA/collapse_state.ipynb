```python
import pandas as pd
```

```python
ca_file = r'../../data/demographics/cc-est2019-alldata-06.csv'
ca = pd.read_csv(ca_file)
ca.head()
```

```python
fl_file = r'../../data/demographics/cc-est2019-alldata-12.csv'
fl = pd.read_csv(fl_file)
fl.head()
```

```python
ga_file = r'../../data/demographics/cc-est2019-alldata-13.csv'
ga = pd.read_csv(ga_file)
ga.head()
```

```python
ny_file = r'../../data/demographics/cc-est2019-alldata-36.csv'
ny = pd.read_csv(ny_file)
ny.head()
```

```python
def collapse_file(df):
    df = df.drop(['SUMLEV', 'STATE', 'COUNTY', 'CTYNAME'], axis=1)
    df = df[(df['YEAR']==1) | (df['YEAR']==12)]
    age_desc = {0: 'Total'
    ,1: '0-4'
    ,2 : '5-9'
    ,3 : '10-14'
    ,4 : '15-19'
    ,5 : '20-24'
    ,6 : '25-29'
    ,7 : '30-34'
    ,8 : '35-39'
    ,9 : '40-44'
    ,10 : '45-49'
    ,11 : '50-54'
    ,12 : '55-59'
    ,13 : '60-64'
    ,14 : '65-69'
    ,15 : '70-74'
    ,16 : '75-79'
    ,17 : '80-84'
    ,18 : '85_plus'}
    df['AGEGRP'] = df['AGEGRP'].replace(age_desc)
    year_desc = {1: '2010 Census', 12: '2019 Estimate'}
    df['YEAR'] = df['YEAR'].replace(year_desc)
    rst = df.groupby(['STNAME', 'YEAR', 'AGEGRP']).sum().reset_index()
    return rst
```

```python
out_ny = collapse_file(ny)
out_ca = collapse_file(ca)
out_fl = collapse_file(fl)
out_ga = collapse_file(ga)
```

```python
path = '../../data/demographics/'
out_ny.to_csv(path + 'NY_ST_Demo.csv', index=False)
out_ca.to_csv(path + 'CA_ST_Demo.csv', index=False)
out_fl.to_csv(path + 'FL_ST_Demo.csv', index=False)
out_ga.to_csv(path + 'GA_ST_Demo.csv', index=False)
```
