```python
import pandas as pd
import c3aidatalake
```

```python
def get_JHU(state):
    today = pd.Timestamp.now().strftime("%Y-%m-%d")
    rst = c3aidatalake.evalmetrics(
        "outbreaklocation",
        {
            "spec" : {
                "filter": "contains(id, '{}')".format(state),
                #"filter": """contains(id, 'Georgia') || contains(id, 'Florida') 
                #            || contains(id, 'California') || contains(id, 'NewYork')""",
                "expressions" : ['JHU_ConfirmedCases', 'JHU_ConfirmedDeaths', 'JHU_ConfirmedRecoveries'],
                "start" : '2020-01-01',
                "end" : today,
                "interval" : "DAY",
            }
        }
    )
    rst['dates'] = rst['dates'].astype(str)
    out = rst.T.reset_index()
    out.columns = out.iloc[0]
    out = out.iloc[1:]
    #out.insert(0, 'measure', out['dates'].apply(lambda x: x.split('UnitedStates.')[1]+'.'+x.split('UnitedStates.')[2]))
    out.insert(0, 'measure', out['dates'].apply(lambda x: 'JHU' + x.split('JHU')[1]))
    #out.insert(0, 'state', state)
    out.drop(['dates'], axis=1, inplace=True)
    #out = out.groupby(['state', 'measure']).sum().reset_index()
    out = out.groupby(['measure']).sum().reset_index()
    out['measure'] = out.apply(lambda row: state + '.' + row['measure'], axis=1)
    ts = out.T
    ts.columns = ts.iloc[0]
    ts = ts[1:].reset_index()
    ts.rename({0:'date'}, axis=1,inplace=True)
    return ts
    #return out
```

```python
ga = get_JHU('Georgia')
ga
```

```python
ca = get_JHU('California')
ca
```

```python
fl = get_JHU('Florida')
fl
```

```python
ny = get_JHU('NewYork')
ny
```

```python
path = '../../data/cases/'
ga.to_csv(path + 'GA_JHU_cases.csv', index=False)
ca.to_csv(path + 'CA_JHU_cases.csv', index=False)
fl.to_csv(path + 'FL_JHU_cases.csv', index=False)
ny.to_csv(path + 'NY_JHU_cases.csv', index=False)
```
